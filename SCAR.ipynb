##### CALCULO RATIO

```python
def calcular_ratio(preferencias, items, user, historico):
    historico_usuario = historico.loc[historico[0]==user]
    if not historico_usuario.empty:
        print("Usuario existente")
    else:
        print("Usuario nuevo")
    
    items_vistos = historico_usuario[1].values # Lista peliculas vistas
    items_no_vistos = items[~items[0].isin(items_vistos)] # Peliculas no vistas
    items_no_vistos_generos = items_no_vistos.iloc[:,1:19] # Peliculas segun generos
    titulos_no_vistos = items_no_vistos.iloc[:,20]
    # Multiplicar el vector preferencias por cada fila de items_no_vistos_generos y sumar el resultado y guardarlo en una lista
    resultados = []
    for index, row in items_no_vistos_generos.iterrows():
        resultados.append(np.dot(preferencias, row))

    # Extraer los 6 índices de las películas con mayor puntuación
    indices = np.argsort(resultados)[-6:]
    # Extraer los títulos de las películas con mayor puntuación
    recomendaciones = titulos_no_vistos.iloc[indices]

    
    return recomendaciones
```

##### CLUSTERING - SEGUNDA APROXIMACIÓN

En este caso obtendremos una especie de embedding en base a las preferencias del usuario que posteriormente usaremos para realizar el clustering

```python
def crear_embedding(historico, items):
    users = historico[0].unique() # Lista de usuarios
    embeddings = []
    for user in users:
        historico_usuario = historico.loc[historico[0]==user]
        items_vistos_usuario = historico_usuario[1].values
        items_vistos_usuario_generos = items[items[0].isin(items_vistos_usuario)].iloc[:,1:19]
        valoraciones_usuario = historico_usuario[2].values
        num_vistas_genero = np.sum(items_vistos_usuario_generos, axis=0)
        preferencias = np.sum(items_vistos_usuario_generos * valoraciones_usuario[:, None], axis=0) / num_vistas_genero
        preferencias = np.nan_to_num(preferencias)
        embeddings.append(list(preferencias))
    
    return np.array(embeddings)
```

```python
print(crear_embedding(u1_base, items))
```

```python
# Build a DataFrame with the embeddings (one per column) and the user id (first column)
embeddings = pd.DataFrame(crear_embedding(u1_base, items))
embeddings.insert(0, 'user', u1_base[0].unique())
embeddings[0] = pd.Categorical(embeddings[0])
```

```python
# Eliminar primera columna
embeddings = embeddings.drop(embeddings.columns[0], axis=1)
print(embeddings)
```

```python
# Método elbow para identificar el número óptimo de clusters
# ==============================================================================
range_n_clusters = range(1, 9)
inertias = []

for n_clusters in range_n_clusters:
    modelo_kmeans = KMeans(
                        n_clusters   = n_clusters, 
                        n_init       = 20, 
                        random_state = 123
                    )
    modelo_kmeans.fit(embeddings)
    inertias.append(modelo_kmeans.inertia_)

fig, ax = plt.subplots(1, 1, figsize=(6, 3.84))
ax.plot(range_n_clusters, inertias, marker='o')
ax.set_title("Evolución de la varianza intra-cluster total")
ax.set_xlabel('Número clusters')
ax.set_ylabel('Intra-cluster (inertia)')
```

```python
range_n_clusters = range(2, 15)
valores_medios_silhouette = []

for n_clusters in range_n_clusters:
    modelo_kmeans = KMeans(
                        n_clusters   = n_clusters, 
                        n_init       = 20, 
                        random_state = 123
                    )
    cluster_labels = modelo_kmeans.fit_predict(embeddings)
    silhouette_avg = silhouette_score(embeddings, cluster_labels)
    valores_medios_silhouette.append(silhouette_avg)
    
fig, ax = plt.subplots(1, 1, figsize=(6, 3.84))
ax.plot(range_n_clusters, valores_medios_silhouette, marker='o')
ax.set_title("Evolución de media de los índices silhouette")
ax.set_xlabel('Número clusters')
ax.set_ylabel('Media índices silhouette')
```

```python
def plot_dendrogram(model, **kwargs):
    '''
    Esta función extrae la información de un modelo AgglomerativeClustering
    y representa su dendograma con la función dendogram de scipy.cluster.hierarchy
    '''
    
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot
    dendrogram(linkage_matrix, **kwargs)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
altura_corte = 6
plot_dendrogram(modelo_hclust_ward, color_threshold=altura_corte, ax=ax)
ax.set_title("Distancia euclídea, Linkage ward")
ax.axhline(y=altura_corte, c = 'black', linestyle='--', label='altura corte')
ax.legend()
```
